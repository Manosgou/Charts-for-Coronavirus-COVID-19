The API which i used is provided from this [website](https://thevirustracker.com/),which includes also [documentation](https://thevirustracker.com/api).
By simply replacing the ```country``` code which is stored in the variable named country,you can receive all of the data for a specific Country.The full country code list is [here](https://thevirustracker.com/api#indexpage).

````
country='GR'
url_stats = f'https://api.thevirustracker.com/free-api?countryTotal={country}'
url_timeline =f'https://api.thevirustracker.com/free-api?countryTimeline={country}'


response_stats = requests.get(url_stats)
response_timeline = requests.get(url_timeline)


stats = response_stats.json()
timeline = response_timeline.json()
````
